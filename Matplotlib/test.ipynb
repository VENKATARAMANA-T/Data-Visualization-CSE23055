```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]                 # Data set
y = [2,3,4,5,6]

plt.plot(x,y,marker='o',linestyle="-",color="blue")        # used to desing the plot

plt.title("Sample Plot")        # used to give title
plt.xlabel("X-axis")            # used to give X-Label
plt.ylabel("Y-axis")            # used to give Y-Label
plt.legend(['Datapoints'],loc='upper left')


plt.show()                      # used to show plot
# plt.savefig('test1.png')     # used to save image of plot  (Use savefig or show only do not use both at a time.)
```
